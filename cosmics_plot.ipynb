```python
"""
Read and study the CORSIKA database
"""

from cosmics import *
from ROOT import TFile, TH1D, TH2D
```

```python
#variable declaration

# location of the database at CERN
directory = '/eos/experiment/wa105/offline/LArSoft/MC/CORSIKADB/standard/'

# sqlite query
query = 'SELECT pdg, e, px, py, pz, (px*px+py*py+pz*pz) from particles;'

#output rootfile
ofile = TFile('corsika_hist.root', 'RECREATE')

#particles code to process
pdg_codes = [13, 11, 2112, 211] #[muons, electrons, protons, pioins]
```

```python
#fetch info from the database and fill
db_array = readdb(directory, query)
```

```python
#get particles species arrays (ignore particles-antiparticles differences for the moment)

#muons_array = db_array[np.where( abs(db_array[:,0]) == 13 )]
#electrons_array = db_array[np.where( abs(db_array[:,0]) == 11 )]
#proton_array = db_array[np.where( abs(db_array[:,0]) == 2112 )]
#pion_array = db_array[np.where( abs(db_array[:,0]) == 211 )]
```

```python
#loop over the particle codes 

for pdg in pdg_codes:
    
    print "processing particle "+str(pdg)

    #define historgrams
    hEnergy = TH1D("hEnergy"+str(pdg), str(pdg)+"_Energy;Energy (GeV)", 100, 0, 6)
    hTheta = TH1D("hTheta"+str(pdg), str(pdg)+"_Theta;#theta", 100, 0, 2*np.pi)
    hPhi = TH1D("hPhi"+str(pdg), str(pdg)+"_Phi;#phi (rad)", 100, -2*np.pi, 2*np.pi)
    hEnergyTheta = TH2D("hEnergyTheta"+str(pdg), str(pdg)+" energy vs az. angle (CORSIKA);Energy (GeV);#theta ", 100, 0, 20, 100, 0, np.pi ) #for the moment with cos(theta)
    hEnergyPhi = TH2D("hEnergyPhi"+str(pdg), str(pdg)+" energy vs horiz. angle (CORSIKA);Energy (GeV);#phi (rad)", 100, 0, 20, 100, -np.pi, np.pi )

    #select species array
    particle_array = db_array[np.where( abs(db_array[:,0]) == pdg )]
    
    #fill the histograms
    fill_hist2D( hEnergyTheta, particle_array[:,1], particle_array[:,2]  )
    fill_hist2D( hEnergyPhi  , particle_array[:,1], particle_array[:,3]  )
    fill_hist1D( hEnergy, particle_array[:,1] );
    fill_hist1D( hTheta, particle_array[:,2] )
    fill_hist1D( hPhi, particle_array[:,3] )
    
    #writhe in the output file
    ofile.Write()
```

```python
#close the output file
ofile.Close()
```
